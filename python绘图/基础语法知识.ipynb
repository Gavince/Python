```python
import matplotlib.pyplot as plt
import numpy as np

def f(t):
    """显示cos函数图像"""
    
    return np.exp(-t)*np.cos(2 * np.pi * t)


a = np.arange(0.0, 5.0, 0.02)

#1. 绘制子区域１
plt.subplot(221)
plt.plot(a, f(a))
#2. 绘制子区域２
plt.subplot(222)
plt.plot(a, np.cos(2 * np.pi * a), 'r--')

plt.show()
plt.savefig("Image_2", dpi = 620)
```
