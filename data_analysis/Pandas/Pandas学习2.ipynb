### 2.4 数据透视

```python
#查看所有数据
pd.set_option('max_columns', 100)
pd.set_option('max_rows', 500)
```

```python
#1.基本形式
pd.pivot_table(df, index=["年代"])#指定年代来来进行数据分析,计算平均数
```

```python
#2.多个索引(默认平均值)
pd.pivot_table(df, index=["年代", '产地'], values=["投票人数"],)
```

```python
#4.指定特定函数
pd.pivot_table(df, index=['年代', '产地'], values=["投票人数"], aggfunc=np.sum)
```

```python
#多函数统计
pd.pivot_table(df, index=["产地"], values=["投票人数", "评分"], aggfunc=[np.sum, np.mean])
```

```python
#非数值处理, 使用fill_value使得nan为特定值
pd.pivot_table(df, index=["产地"], aggfunc=[np.sum, np.mean], fill_value=0)
```

```python
#汇总统计
pd.pivot_table(df, index=["产地"], aggfunc=[np.sum, np.mean], fill_value=0, margins=True)
```

```python
#对特定列进行数据统计
pd.pivot_table(df, index=["产地"], aggfunc={"投票人数":np.sum, "评分":np.mean}, fill_value=0)
```
