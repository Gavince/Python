## 基本方法掌握

```python
#用来显示包文件
import matplotlib
matplotlib.matplotlib_fname()
```

```python
import matplotlib.pyplot as plt


x = np.arange(1, 100)
y = np.sin(x)
# line1, line2 = plt.plot(x, y,'r--', x, y+1, 'bo' )
plt.figure(figsize=(10, 10))#设置画布大小值
# plt.setp(line1, color="g", linewidth = 4)#设置具体的设定
# plt.savefig("Data")

#绘制第一个图像
plt.subplot(211)
plt.plot(x, y,"ro" ,label="data")
plt.title("sin")
plt.xlim(1, 99)
plt.ylim(-2, 3)
plt.legend()

#绘制第二个图像
plt.subplot(212)
plt.plot(x, np.cos(x),"g*", label="data")
plt.title("cos")
plt.xlim(1, 99)
plt.ylim(-2, 3)
plt.legend()

plt.show()
```

## 电影数据分析

### (1) 柱状图

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

```

```python
df = pd.read_excel("douban.xlsx")
df.名字[:20]
```

```python
df["年代"][:5]#绘制前五行数据
```

```python
data = df["产地"].value_counts()
data
```

```python
x = data.index
y = data.values

#设置画布大小
plt.figure(figsize=(15, 10))

#绘制数据
plt.bar(x, y, color="g")
plt.title("各个国家或地区电影数量", fontsize = 30)
plt.xlabel("国家/地区", fontsize = 15)
plt.ylabel("电影数量", fontsize = 15)
plt.tick_params(labelsize = 14)
plt.xticks(rotation=90)

for a, b in zip(x, y):
    plt.text(a, b, b+20, ha="center", va="bottom", color="b", fontsize=10)#x轴, y轴， text
#最后绘制出图线
plt.show()
```

### (2) 折线图

```python
data = df["年代"][:4000]
data = data.value_counts()
```

```python
x = data.index
y = data.values

#绘制折线图
plt.subplot(121)
plt.Figure(figsize=(15, 10))
plt.plot(x, y, color = "b")
plt.title("每年的电影数量", fontsize = 10)
plt.xlabel("年份", fontsize = 10)
plt.ylabel("数量", fontsize = 10)

for a, b in zip(x[::10], y[::10]):
    plt.text(a, b+10, b, ha="center", va="bottom", fontsize=10)

    
plt.annotate("2012年达到最大值",xy=(2012, data[2012]))
plt.savefig("电影数量数据")
plt.text(2000, 150, "增长速度最快", ha ="right")

#绘制散点图
plt.subplot(122)
plt.Figure(figsize=(15, 15))
plt.scatter(x, y, color = "b")
plt.title("每年的电影数量", fontsize = 10)
plt.xlabel("年份", fontsize = 10)
plt.ylabel("数量", fontsize = 10)

for a, b in zip(x[::10], y[::10]):
    plt.text(a, b+10, b, ha="center", va="bottom", fontsize=10)

    
plt.annotate("2012年达到最大值",xy=(2012, data[2012]))
plt.savefig("电影数量数据")
plt.text(2000, 150, "增长速度最快", ha ="right")

plt.show()
```

### (3)饼状图

```python
data = pd.cut(df["时长"][:100],[0, 60, 90, 110, 1000]).value_counts()
data
```

```python
y = data.values
# y = y /sum(y)#计算数据所占比例


plt.figure(figsize=(10, 10))
plt.title("电影时长占比",fontsize=10)
plt.pie(y,labels=data.index,autopct="%.lf%%", startangle=90)

plt.legend()
plt.show()
```

### 数据展示2

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

#导入数据
df = pd.read_excel("douban.xlsx")
```

```python
### 双轴线的画法
fig = plt.figure(figsize=(10, 8))
ax1 = fig.add_subplot(111)
n, bins, patch = ax1.hist(df["评分"], bins=100, color="m")

ax1.set_ylabel("电影数量", fontsize = 10)
ax1.set_xlabel("评分", fontsize = 10)
ax1.set_title("频率分布图")

y = mlab.normpdf(bins, df["评分"].mean(),df["评分"].std() )

ax2 = ax1.twinx()
ax2.plot(bins, y,'b--')

plt.show()
```

### (1)散点图

```python
x = df['时长'][:1000]
y = df['评分'][:1000]

plt.figure(figsize=(10, 8))
plt.xlabel("时长",)
plt.ylabel("评分")
plt.title("电影时长和评分散点图")
plt.scatter(x, y, color = "g", marker="p")
plt.legend()
plt.savefig("电影时长和评分散点图")
plt.show()
```
