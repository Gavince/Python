## 基本方法掌握

```python
#用来显示包文件
import matplotlib
matplotlib.matplotlib_fname()
```

```python
x = np.arange(1, 100)
y = np.sin(x)
# line1, line2 = plt.plot(x, y,'r--', x, y+1, 'bo' )
plt.figure(figsize=(10, 10))#设置画布大小值
# plt.setp(line1, color="g", linewidth = 4)#设置具体的设定
# plt.savefig("Data")
plt.subplot(211)
plt.plot(x, y,"ro" ,label="data")
plt.title("sin")
plt.xlim(1, 99)
plt.ylim(-2, 3)
plt.legend()


plt.subplot(212)
plt.plot(x, np.cos(x),"g*", label="data")
plt.title("cos")
plt.xlim(1, 99)
plt.ylim(-2, 3)
plt.legend()

plt.show()
```

## 电影数据分析

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
chinese_font = FontProperties(fname='/usr/share/fonts/MyFonts/YaHei.Consolas.1.11b.ttf')
```

```python
df = pd.read_excel("douban.xlsx")
```

```python
data = df["产地"].value_counts()
data
```

```python
x = data.index
print(x)
y = data.values
plt.figure(figsize=(10, 6))
plt.bar(x, y, color="g")
plt.title("各个国家或地区电影数量", fontsize = 20)
plt.xlabel("国家")
plt.ylabel("电影数量")
plt.show()
```
