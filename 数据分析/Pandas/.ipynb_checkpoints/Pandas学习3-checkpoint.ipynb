### 3.2 数据分组, 分组运算

```python
#Groupby技术
group = df.groupby(df.产地)
```

```python
type(group)
```

```python
#每一年各个场地的均值
group.mean()
```

```python
#总数
group.sum()
```

```python
#计算每年的平均分
df.评分.groupby(df.年代).mean()
```

```python
#注意以上只会对数值进行计算
df.年代 = df.年代.astype("str")
```

```python
df.groupby(df.产地).median()
```

```python
#传入多个分组变量来进行数据分组
df.groupby(["产地", "年代"]).mean()#多个变量
```

```python
#每一个地区,每一个年代
group = df.评分.groupby([df["产地"], df["年代"]]).mean()
```

### 3.离散化处理

```python
#pd.cut 函数
```

```python
#bin 区间, 标签
df.评分等级 = pd.cut(df.评分, [0, 3, 5, 7, 9, 10], labels=["E", "D", "C", "B", "A"])
```

```python
#投票人数的排序
bins = np.percentile(df.投票人数,[0, 20, 40,60, 80, 100])
bins
```

```python
df["热门程度"] =pd.cut(df.投票人数, bins, labels=["E", "D", "C", "B", "A"])
```

```python
df.head(5)
```

```python
df[df.热门程度 == "A"].head(5)
```

```python
#横向的拼接数据集
df1 = df.loc[:5]
df1
```

```python
df2 = df.loc[:5][["名字", "产地"]]
df2["票房"] = [2,1 ,2, 44, 660, 44]
```
