```python
#描述数据
df.评分.mean()
df.评分.max()
df.评分.min()
df.评分.describe()#描述一行数据的额有用信息
df.产地.describe(include=["object"])#如果想要查看非数字类型的列的统计指标，可以设置 include=["object"] 来获得
```

```python
#添加一行数据
dit = {"名字":"钢铁侠", "投票人数":111111, "类型":"科幻","产地":"美国","上映时间":"2018","时长":142,
      "年代":2018, "评分":np.nan, "首映地点":"美国"}
dit
```

```python
s = pd.Series(dit)
s
# s.name = 38738
```

```python
df = df.append(s)#插入一行数据后必须覆盖(必须这样操作)
df.tail(1)
```

```python
df = df.drop(38738)#消除最后一行数据
```

```python
df.tail(1)
```

```python
#列操作
df.columns
```

```python
df["时长"][:5]#展示某一列某几行的数据
```

```python
df = df.drop("序号", axis=1)#需要指定行列的位置数据
df
```

```python
#取得某一数据
df.loc[1,"名字"]
df.loc[[1, 2, 3], ["名字", "评分"]]#取得特定行和列的数据进行处理
```

### 条件选择

```python
df[df["产地"] == "美国"][:5]#取得特定数据
```

```python
df[(df.产地=="美国") & (df.评分 >= 9.0)][:5]#比较语句必须家括号
```

```python
df[((df.产地 == "美国" )|(df.产地 == "中国大陆")) & (df.评分 > 9)]
```
